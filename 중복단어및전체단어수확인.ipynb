```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 엑셀 파일 읽기
file_path = r'C:\Users\scott\work\2023\헌법재판소\헌법재판소.xlsx'
sheet_name = 'Sheet1'
df = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')

# F6부터 F71까지의 데이터 추출
data = df.iloc[5:71, 5]

# 공백으로 단어를 분할
words = [str(word).split() for word in data]

# 전체 단어 수와 중복 단어 수 계산
all_words = []
duplicated_words = []

for word_list in words:
    for word in word_list:
        if word not in all_words:
            all_words.append(word)
        else:
            duplicated_words.append(word)

total_word_count = len(all_words)
duplicated_word_count = len(duplicated_words)

print(f'전체 단어 수: {total_word_count}')
print(f'중복 단어 수: {duplicated_word_count}')

# 중복 단어별 횟수 계산
duplicated_word_counts = Counter(duplicated_words)

# 중복 횟수를 기준으로 단어 정렬
sorted_duplicated_word_counts = sorted(duplicated_word_counts.items(), key=lambda x: x[1], reverse=True)

print("중복 횟수가 많은 순서대로 단어:")
for word, count in sorted_duplicated_word_counts:
    if count > 1:
        print(f'{word}: {count}회')

```
